```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import control
from numpy import sin, cos, tan, pi
from control import TransferFunction as TF
```

```python
plt.rcParams['font.size'] = 14
```

```python
t = np.arange(0,2,0.002)
```

```python
p = 2*pi*5
```

```python
G = TF(p,[1,p])
```

```python
f = 3
```

```python
u = np.sin(2*pi*f*t)
```

```python
to, y = control.forced_response(G,t,u)
```

```python
plt.figure()
plt.plot(t,u,t,y)
plt.xlabel('Time (sec.)')
plt.ylabel('Signal Amp.')
plt.legend(['u','y'],loc=3)
plt.title("f = %0.4g Hz" % f)
plt.tight_layout()
plt.savefig('fixed_sine_example.png',dpi=300)
```

```python
def fixed_sine_sim(f,xlim=None):
    u = np.sin(2*pi*f*t)
    to, y = control.forced_response(G,t,u)

    plt.figure()
    plt.plot(t,u,t,y)
    plt.xlabel('Time (sec.)')
    plt.ylabel('Signal Amp.')
    plt.legend(['u','y'],loc=3)
    plt.title("f = %0.4g Hz" % f)
    if xlim is not None:
        plt.xlim(xlim)
    plt.tight_layout()
```

```python
fixed_sine_sim(10,[1,1.5])
```

```python
fixed_sine_sim(30,[1.8,2])
```
